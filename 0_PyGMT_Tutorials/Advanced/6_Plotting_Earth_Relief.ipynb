# Plotting Earth Relief
- This tutorial can be found at this [link](https://www.pygmt.org/latest/tutorials/advanced/earth_relief.html)
- In this tutorial, we will be learning how to plot a map of Earth Relief

```python
import pygmt
```

```python
# Load sample Earth relief data for the entire globe at a resolution of 1 arc degree
grid = pygmt.datasets.load_earth_relief(resolution="01d")
```

```python
# Create a plot
fig = pygmt.Figure()
fig.grdimage(grid=grid)
fig.show()
```

```python
# Using a "Winkel Tripel" projection for the map
fig = pygmt.Figure()
fig.grdimage(grid=grid, projection="R12c")
fig.show()
```

```python
# Setting the Color Palette Table (CPT) of the figure
fig = pygmt.Figure()
fig.grdimage(grid=grid, projection="R12c", cmap="geo")
fig.show()
```

```python
# Using the relief CPT
fig = pygmt.Figure()
fig.grdimage(grid=grid, projection="R12c", cmap="relief")
fig.show()
```

```python
# Adding a color bar: Displays the CPT & the associated Z-values
fig = pygmt.Figure()
fig.grdimage(grid=grid, projection="R12c", cmap="geo")
fig.colorbar(frame=["a2500", "x+lElevation", "y+lm"])
fig.show()
```

```python
# Creating a region map: Provides data for a specific area
# The `region` argument is required for resolutions at 5 arc minutes or higher
grid = pygmt.datasets.load_earth_relief(resolution="10m", region=[-14, 30, 35, 60])
fig = pygmt.Figure()
fig.grdimage(grid=grid, projection="M15c", frame="a", cmap="geo")
fig.colorbar(frame=["a1000", "x+lElevation", "y+lm"])
fig.show()
```
