# Intro to PyGMT
- This tutorial can be found at this [link](https://www.pygmt.org/latest/get-started/index.html)
- In this tutorial, we will be making our first PyGMT figure.

```python
import pygmt
```

```python
# Creating a Figure
fig = pygmt.Figure()
# Selecting the region for our figure
fig.coast(region = [-69, -68, 43.75, 44.75], shorelines = True)
fig.show()
```

```python
# Coloring the land and water
fig = pygmt.Figure()
fig.coast(
  region = [-69, -68, 43.75, 44.75], shorelines = True,
  land = "lightgreen", water = "lightblue"
)
fig.show()
```

```python
# Setting the Projection
# Here, we are using the Mercator projection, where the width of the figure is 10cm
fig = pygmt.Figure()
fig.coast(
  region = [-69, -68, 43.75, 44.75], shorelines = True,
  land = "lightgreen", water = "lightblue", projection = "M10c"
)
fig.show()
```

```python
# Adding a frame
# "a" denotes that the axes are annotated automatically
fig = pygmt.Figure()
fig.coast(
  region = [-69, -68, 43.75, 44.75], shorelines = True,
  land = "lightgreen", water = "lightblue", projection = "M10c",
  frame = "a"
)
fig.show()
```

```python
# Adding a title
fig = pygmt.Figure()
fig.coast(
  region = [-69, -68, 43.75, 44.75], shorelines = True,
  land = "lightgreen", water = "lightblue", projection = "M10c",
  frame = ["a", "+tMaine"]
)
fig.show()
```
