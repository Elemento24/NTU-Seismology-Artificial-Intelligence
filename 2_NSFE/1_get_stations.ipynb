```python
import os
from obspy.clients.fdsn import Client
from obspy import UTCDateTime
```

```python
# Creating a directory to store all the Seismic stations' info
datadir = "./"
path = os.path.join(datadir, "Stations")

try:
  os.mkdir(path)
except:
  print("The directory already exists")
```

```python
# Define stations provider
provider1 = "IRIS"
provider2 = "GFZ"
client_1 = Client(provider1)
client_2 = Client(provider2)

# Define the study site
minlat = 16
maxlat = 30
minlon = 90
maxlon = 104
```

```python
# Defining the catalog duration
for i in range(22):
  year1 = 2000 + i
  year2 = 2001 + i
  
  print(f"For the Time-period: {year1}-{year2}")
  print(f"For provider: {provider1}")
  
  try:
    ## See details at https://docs.obspy.org/packages/autogen/obspy.clients.fdsn.client.Client.get_stations.html
    inv = client_1.get_stations(
      minlatitude = minlat-0.5, maxlatitude = maxlat+0.5, minlongitude = minlon-0.5, maxlongitude = maxlon+0.5,
      channel = "BH?,HH?,EH?", level = "channel", starttime = UTCDateTime(year1, 1, 1, 0, 0), 
      endtime = UTCDateTime(year2, 1, 1, 0, 0)
    )

    print("A total of {} networks~".format(len(inv)))

    # Write to a text file
    inv.write(f"{path}/stations_{provider1}_{year1}-{year2}.txt", format="STATIONTXT")
  except:
    print(f"{provider1} doesn't have data for this particular region and duration.")
  
  print(f"For provider: {provider2}")
  
  try:
    ## See details at https://docs.obspy.org/packages/autogen/obspy.clients.fdsn.client.Client.get_stations.html
    inv = client_2.get_stations(
      minlatitude = minlat-0.5, maxlatitude = maxlat+0.5, minlongitude = minlon-0.5, maxlongitude = maxlon+0.5,
      channel = "BH?,HH?,EH?", level = "channel", starttime = UTCDateTime(year1, 1, 1, 0, 0), 
      endtime = UTCDateTime(year2, 1, 1, 0, 0)
    )

    print("A total of {} networks~".format(len(inv)))

    # Write to a text file
    inv.write(f"{path}/stations_{provider2}_{year1}-{year2}.txt", format="STATIONTXT")
  except:
    print(f"{provider2} doesn't have data for this particular region and duration.")
    
  print()
```
