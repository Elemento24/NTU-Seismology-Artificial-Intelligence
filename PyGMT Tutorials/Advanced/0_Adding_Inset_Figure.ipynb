# Adding an inset to the figure
- This tutorial can be found at this [link](https://www.pygmt.org/latest/tutorials/advanced/insets.html)
- In this tutorial, we will be plotting an inset figure inside another larger figure.

```python
import pygmt
```

```python
# Creating the Larger Figure
fig = pygmt.Figure()
fig.coast(
    region=[-74, -69.5, 41, 43],
    borders="2/thin", shorelines="thin", projection="M15c",
    land="lightyellow", water="lightblue", frame="a",
)
fig.show()
```

```python
# Creating an inset figure within the larger figure
fig = pygmt.Figure()
fig.coast(
    region=[-74, -69.5, 41, 43],
    borders="2/thin", shorelines="thin", projection="M15c",
    land="lightyellow", water="lightblue", frame="a",
)

# Position includes one of the nine anchor positions, in this case, "BL"
with fig.inset(position="jBL+w3c", box="+pblack+glightred"):
    # Pass is used to exit the with statement as no plotting functions are called
    pass
fig.show()
```

```python
fig = pygmt.Figure()
fig.coast(
    region=[-74, -69.5, 41, 43],
    borders="2/thin", shorelines="thin", projection="M15c",
    land="lightyellow", water="lightblue", frame="a",
)

# Defining the offset of the inset figure using "+o" for X and Y axis respectively
with fig.inset(position="jBL+w3c+o0.5c/0.2c", box="+pblack+glightred"):
    pass
fig.show()
```

```python
# Creating an inset figure within the larger figure
fig = pygmt.Figure()
fig.coast(
    region=[-74, -69.5, 41, 43],
    borders="2/thin", shorelines="thin", projection="M15c",
    land="lightyellow", water="lightblue", frame="a",
)

# Calling Standard Plotting functions from within the `inset` context manager
# This does not include an inset fill as it is covered by the inset figure
with fig.inset(position="jBL+w3c+o0.5c/0.2c", box="+pblack"):
  fig.coast(
    region=[-80, -65, 35, 50], projection="M3c",
    land="gray", water="white", borders=[1, 2], shorelines="1/thin",
    # Use dcw to selectively highlight an area
    dcw="US.MA+gred",
  )
fig.show()
```
