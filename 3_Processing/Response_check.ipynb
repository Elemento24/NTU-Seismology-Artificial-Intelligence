```python
import glob
import os
from obspy import read_inventory, UTCDateTime
from obspy.clients.fdsn import Client
```

```python
provider = "GFZ" 
year1 = 2015
year2 = 2016
root_dir = "/media/elemento/Element/NTU_RW/NEFS"
data_staxml = f'{root_dir}/Download/{provider}/stations/{year1}'

startDate=f"{year1}-01-01T00:00:00" 
endDate=f"{year2}-01-01T00:00:00"

if not os.path.exists(data_staxml):
    print("No staxml files found!\n")
```

```python
### Ensure the response file compatible with mseed file, i.e., the same recording duration
for xml in glob.glob(f"{data_staxml}/*.xml"): 
    print(xml)
    inv = read_inventory(xml)
    
    # Network
    inv[0].start_date = startDate
    inv[0].end_date = endDate
    
    # Station
    inv[0][0].start_date = startDate
    inv[0][0].end_date = endDate
    
    # Channel
    nc = len(inv[0][0])
    for i in range(nc):
        inv[0][0][i].start_date = startDate
        inv[0][0][i].end_date = endDate
    
    # Save the modified file
    inv.write(xml, format="stationxml", validate=True)
```

```python
### Reobtain the response file

# Specify webservice
client = Client(provider)
starttime = UTCDateTime(f"{year1}-01-01")
endtime = UTCDateTime(f"{year2}-01-01")

# Specify network, station and channel
net = "IS"
sta = "MDBI"
channel = "BHN,BHE,BHZ"

inv = client.get_stations(network=net, station=sta, starttime=starttime, endtime=endtime, channel=channel, level="response")
print(inv)
```

```python
# Network
inv[0].start_date = startDate
inv[0].end_date = endDate

# Station
inv[0][0].start_date = startDate
inv[0][0].end_date = endDate

# Channel
nc = len(inv[0][0])
for i in range(nc):
    inv[0][0][i].start_date = startDate
    inv[0][0][i].end_date = endDate
    
# Write to a STATIONXML file  
name = "{}.{}".format(net,sta)
inv.write(f"{data_staxml}/{name}.xml", format="STATIONXML")
```
