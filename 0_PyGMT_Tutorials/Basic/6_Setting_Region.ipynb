# Setting the Region
- This tutorial can be found at this [link](https://www.pygmt.org/latest/tutorials/basics/regions.html)
- In this tutorial, we will be setting the region.

```python
import pygmt
```

```python
# Set the x-range from 10E to 20E and the y-range to 35N to 45N
# Passing the region in the form of a string
fig = pygmt.Figure()
fig.coast(
  region="10/20/35/45",
  projection="M15c", land="lightgray", water="white",
  borders="1/0.5p", shorelines="1/0.5p", frame="ag",
)
fig.show()
```

```python
# Set the x-range from 10E to 20E and the y-range to 35N to 45N
# Passing the region in the form of a list
fig = pygmt.Figure()
fig.coast(
  region=[10, 20, 35, 45],
  projection="M12c", land="lightgray", water="white",
  borders="1/0.5p", shorelines="1/0.5p", frame="ag",
)
fig.show()
```

```python
# Set the x-range from 10E to 20E and the y-range to 35N to 45N
# Passing the region in the form of corners' coordinates
fig = pygmt.Figure()
fig.coast(
  region="10/35/20/45+r",
  projection="M12c", land="lightgray", water="white",
  borders="1/0.5p", shorelines="1/0.5p", frame="ag",
)
fig.show()
```

```python
# Setting the Global Regions, "d"
# In this case, the range is 180W to 180E (-180, 180) and 90S to 90N (-90 to 90)
fig = pygmt.Figure()
fig.coast(
  region = "d", projection = "Cyl_stere/12c", land="lightgray", water="white",
  borders="1/0.5p", shorelines="1/0.5p", frame="ag",
)
fig.show()
```

```python
# Setting the Global Regions, "g"
# In this case, the range is 0E to 360E (0, 360) and 90S to 90N (-90 to 90)
fig = pygmt.Figure()
fig.coast(
  region = "g", projection = "Cyl_stere/12c", land="lightgray", water="white",
  borders="1/0.5p", shorelines="1/0.5p", frame="ag",
)
fig.show()
```

```python
# Setting Regions via ISO Code
fig = pygmt.Figure()
fig.coast(
  region = "JP", projection = "M12c", land="lightgray", water="white",
  borders="1/0.5p", shorelines="1/0.5p", frame=["ag", "+tJapan"],
)
fig.show()
```

```python
# Expanding the regions defined by ISO Code
# Here, the increment is by 3 degrees uniformly in all the directions
fig = pygmt.Figure()
fig.coast(
  region = "JP+r3", projection = "M12c", land="lightgray", water="white",
  borders="1/0.5p", shorelines="1/0.5p", frame=["ag", "+tJapan"],
)
fig.show()
```

```python
# Expanding the regions defined by ISO Code
# Here, the increment is by 3 degrees on X-axis, and by 5 degrees on Y-axis
fig = pygmt.Figure()
fig.coast(
  region = "JP+r3/5", projection = "M12c", land="lightgray", water="white",
  borders="1/0.5p", shorelines="1/0.5p", frame=["ag", "+tJapan"],
)
fig.show()
```

```python
# Expanding the regions defined by ISO Code
# Here, the increment is done differently in all the 4 directions
fig = pygmt.Figure()
fig.coast(
  region = "JP+r3/5/7/9", projection = "M12c", land="lightgray", water="white",
  borders="1/0.5p", shorelines="1/0.5p", frame=["ag", "+tJapan"],
)
fig.show()
```

```python
# Expanding the regions defined by ISO Code
# Here, the increment is by 3 degrees uniformly in all the directions
# It adds the increment without rounding
fig = pygmt.Figure()
fig.coast(
  region = "JP+R3", projection = "M12c", land="lightgray", water="white",
  borders="1/0.5p", shorelines="1/0.5p", frame=["ag", "+tJapan"],
)
fig.show()
```

```python
# Expanding the regions defined by ISO Code
# Expands the final region boundaries to be multiples of increment
# Ensures that the bounding box extends by at least 0.25 times the increment
fig = pygmt.Figure()
fig.coast(
  region = "JP+e3", projection = "M12c", land="lightgray", water="white",
  borders="1/0.5p", shorelines="1/0.5p", frame=["ag", "+tJapan"],
)
fig.show()
```
