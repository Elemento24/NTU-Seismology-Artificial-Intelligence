# Creating a map with contour lines
- This tutorial can be found at this [link](https://www.pygmt.org/latest/tutorials/advanced/contour_map.html)
- In this tutorial, we will be creating a map with contour lines

```python
import pygmt
```

```python
# Load sample earth relief data
grid = pygmt.datasets.load_earth_relief(resolution="05m", region=[-92.5, -82.5, -3, 7])
```

```python
# Creating a contour plot with annotated and unannotated contour lines
# Default contour line intervals are 500 meters, with an annotated contour line every 1000 meters
fig = pygmt.Figure()
fig.grdcontour(grid=grid)
fig.show()
```

```python
# Adjusting the contour line settings
# There are contour intervals every 250 meters and annotated contour lines every 1000 meters
fig = pygmt.Figure()
fig.grdcontour(
  annotation = 1000, interval = 250, grid = grid
)
fig.show()
```

```python
# Defining the contour limits
# We can define the limit as either a list or a string
fig = pygmt.Figure()
fig.grdcontour(
  annotation = 1000, interval = 250, grid = grid,
  limit = [-4000, -2000]
)
fig.show()
```

```python
# Adding map settings to contour plots
fig = pygmt.Figure()
fig.grdcontour(
  annotation = 1000, interval = 250, grid = grid,
  limit = [-4000, -2000], projection = "M10c", frame = True
)
fig.show()
```

```python
# Adding a colormap to the contour map
# Must be called prior to `grdcontour` to keep the contour lines visible
fig = pygmt.Figure()
fig.grdimage(
  grid = grid, cmap = "haxby", projection = "M10c", frame = True
)
fig.grdcontour(
  annotation = 1000, interval = 250, grid = grid,
  limit = [-4000, -2000]
)
fig.show()
```
