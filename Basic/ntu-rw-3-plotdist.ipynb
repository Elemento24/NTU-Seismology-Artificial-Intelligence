# Third Step | Plot Distributions
- Make the plot to show the distribution of earthquakes and seismic stations in our study region.

### References
- [PyGMT Documentation](https://www.pygmt.org/latest/index.html)
- [Intro to PyGMT](https://www.pygmt.org/latest/get-started/first_figure.html)

```python
import pygmt
import pandas as pd
import numpy as np
```

```python
# Define the Input Directory
datadir = "./"

# Define the min-max latitudes and longitudes for plotting the map
minlon, maxlon = 92, 98
minlat, maxlat = 15.5, 27.5
```

- The below file is a remote data set stored on one of the several remote global data grids offered by **GMT**.
- For more information, refer to this [link](https://docs.generic-mapping-tools.org/6.1/datasets/remote-data.html)

```python
# Define etopo data file
# This file consists of "2-D gridded data set or image to be plotted"
topo_data = '@earth_relief_30s' #30 arc second global relief (SRTM15+V2.1 @ 1.0 km)
```

```python
# Creating a PyGMT figure
fig = pygmt.Figure()

# Configuring the PyGMT
## Reference: https://docs.generic-mapping-tools.org/latest/gmt.conf.html#map-parameters
pygmt.config(MAP_FRAME_TYPE="fancy", MAP_DEGREE_SYMBOL="degree", MAP_FRAME_WIDTH="0.08c",
    MAP_FRAME_PEN="0.1p",MAP_TICK_PEN_PRIMARY="0.1p",MAP_TICK_LENGTH_PRIMARY="0.1c")
## https://docs.generic-mapping-tools.org/latest/gmt.conf.html#format-parameters
pygmt.config(FORMAT_GEO_MAP="dddD")
## https://docs.generic-mapping-tools.org/latest/gmt.conf.html#font-parameters
pygmt.config(FONT_LABEL="8p,4", MAP_LABEL_OFFSET="0.1c", FONT_ANNOT_PRIMARY="8p,4",
    FONT_TITLE="10p,4",MAP_TITLE_OFFSET="0.1c")
## https://docs.generic-mapping-tools.org/latest/gmt.conf.html#postscript-parameters
pygmt.config(PS_MEDIA="A4")

## Reference: https://www.pygmt.org/latest/api/generated/pygmt.Figure.grdimage.html
## https://www.pygmt.org/latest/projections/cyl/cyl_mercator.html
fig.grdimage(
    grid=topo_data, region=[minlon, maxlon, minlat, maxlat], cmap="globe",
    projection='M3i', shading=True,
    frame=['xa1f0.5+l"longitude"', 'ya1f0.5+l"latitude"', 'WsNe']
)

## Reference: https://www.pygmt.org/latest/api/generated/pygmt.Figure.coast.html
fig.coast(frame=False, shorelines=True, borders='1/1p,black')

# Plot stations
## Reference: https://www.pygmt.org/latest/api/generated/pygmt.Figure.plot.html
df = pd.read_csv(f"{datadir}/stations.txt", sep='|', usecols=[0,1,4,5])
df.columns = ["net","station","lat","lon"]
fig.plot(x = df["lon"], y = df["lat"], no_clip = True, style = "t0.3c", 
    label='"Seismic Station"')
fig.legend(position="jbl+o0.1c/0.1c", box="+gsnow")

# Creating a figure inside another figure
## Reference: https://www.pygmt.org/latest/api/generated/pygmt.Figure.inset.html
with fig.inset(position="jtr+w3.0c+o0.04c/0.01c"):
    fig.coast(
        region="g",
        projection="G95/22/?",
        land="black",
        water="lightblue"
    )

    # Plot events
    df = pd.read_csv(f"{datadir}/events.csv", sep=',', usecols=[0,1,2,3,5])
    df.columns=["time","lat","lon","dep","mag"]
    fig.plot(x=df["lon"], y=df["lat"], color="brown", style="c0.07c")
    
    # Study area
    fig.plot(
        x=[92, 92, 98, 98, 92],
        y=[15.5, 27.5, 27.5, 15.5, 15.5], 
        pen="0.3p,red"
    )

fig.show()
```
